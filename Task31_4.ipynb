```
%matplotlib inline
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x)/x - 1/x +1))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-2,8) 
plt.grid(True)
plt.show()

from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return ( x**2 - 1-y, np.exp(x)/x - 1/x +1-y)

x1, y1 =  fsolve(equations, (1, 1))
              
print (x1, y1)
```
